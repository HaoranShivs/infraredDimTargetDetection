```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 读取图像
path = "W:/DataSets/ISTD/IRSTD-1k/trainval/images/XDU309.png"
image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

# 检查图像是否正确读取
if image is None:
    print("Error: Could not read image.")
else:
    # 创建CLAHE对象 (参数: clipLimit, tileGridSize)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))

    # 应用CLAHE
    cl1 = clahe.apply(image)

    # 显示原始图像和处理后的图像
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))
    
    axs[0].imshow(image, cmap='gray')
    axs[0].set_title('Original Image')
    axs[0].axis('off')

    axs[1].imshow(cl1, cmap='gray')
    axs[1].set_title('CLAHE Enhanced Image')
    axs[1].axis('off')

    plt.show()
```

```python
# 检查图像是否正确读取
if image is None:
    print("Error: Could not read image.")
else:
    # 高斯模糊
    blurred = cv2.GaussianBlur(image, (0, 0), sigmaX=10, sigmaY=10)

    # 计算细节图像
    detail = cv2.subtract(image, blurred)

    # 增强局部对比度
    enhanced = cv2.addWeighted(image, 1.5, detail, 1.5, 0)

    # 确保像素值在0-255范围内
    enhanced = np.clip(enhanced, 0, 255).astype(np.uint8)

    # 显示原始图像和处理后的图像
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))
    
    axs[0].imshow(image, cmap='gray')
    axs[0].set_title('Original Image')
    axs[0].axis('off')

    axs[1].imshow(enhanced, cmap='gray')
    axs[1].set_title('Locally Enhanced Image')
    axs[1].axis('off')

    plt.show()
```
